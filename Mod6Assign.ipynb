```python
#Use multiprocessing to create three separate processes. Make each one wait a random number of seconds between zero and one, print the current time, and then exit.

import multiprocessing as mp
import time
import random
from datetime import datetime

def get_now():
    return datetime.now().strftime("%H:%M:%S")

def walker(windows, next_queue):
    for window in windows:
        time.sleep(random.random())  
        print(f"[{get_now()}] Walking to {window}")
        next_queue.put(window)

def washer(in_queue, next_queue):
    while True:
        window = in_queue.get()
        time.sleep(random.random())  
        print(f"[{get_now()}] Spraying {window}")
        next_queue.put(window)
        in_queue.task_done()

def wiper(in_queue):
    while True:
        window = in_queue.get()
        time.sleep(random.random())  
        print(f"[{get_now()}] Wiping {window}")
        in_queue.task_done()

if __name__ == "__main__":
    walk_to_wash = mp.JoinableQueue()
    wash_to_wipe = mp.JoinableQueue()
    
    windows = ['front window', 'back window', 'left window', 'right window']

    wiper_proc = mp.Process(target=wiper, args=(wash_to_wipe,), daemon=True)
    wiper_proc.start()

    washer_proc = mp.Process(target=washer, args=(walk_to_wash, wash_to_wipe), daemon=True)
    washer_proc.start()

    walker(windows, walk_to_wash)

    walk_to_wash.join()
    wash_to_wipe.join()
    
    print(f"[{get_now()}] All done!")
```
