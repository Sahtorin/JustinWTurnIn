```python
#4.1 answers
import random

secret = random.randint(1,10)
while True:
    guess = int(input("Guess a number between 1 and 10: "))
    if guess > secret:
        print("Too high")
    elif guess < secret:
        print("Too low")
    elif guess == secret:
        print("Just right")
        break
```
